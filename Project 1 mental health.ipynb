```python
# Import dependencies 
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('resource/2019 mental health data.csv')

# Handle missing values and normalize gender entries
data['What is your gender?'] = data['What is your gender?'].str.strip().str.lower().replace({
    'm': 'male', 
    'male': 'male',
    'f': 'female',
    'female': 'female',
    'other': 'other'
    
})


# Drop rows with missing gender information
data = data.dropna(subset=['What is your gender?'])

# Filter for only male and female genders
data = data[data['What is your gender?'].isin(['male', 'female'])]

# Group by gender and calculate counts
gender_counts = data['What is your gender?'].value_counts()

# Calculate percentages
total_count = gender_counts.sum()
gender_percentages = (gender_counts / total_count) * 100

# Print gender percentages
print("Gender Percentages with Mental Health Issues in Tech Industry:")
print(gender_percentages)

# Visualize the results using a bar plot
gender_percentages.plot(kind='bar', color=['blue', 'pink'], title='Gender Distribution with Mental Health Issues in Tech Industry')
plt.ylabel('Percentage')
plt.xlabel('Gender')
plt.xticks(rotation=0)  
plt.show()

# Visualize the results using a pie plot
gender_percentages.plot(
    kind='pie',
    colors=['blue', 'pink'],
    labels=gender_percentages.index,
    autopct='%1.1f%%',  # Add percentages to slices
    title='Gender Distribution with Mental Health Issues in Tech Industry'
)
plt.ylabel('')  # Remove y-axis label for a cleaner pie chart
plt.show()
```
