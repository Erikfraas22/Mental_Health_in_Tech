```python
# Import dependencies 
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('2019 Mental Health Survey.csv')

# Normalize and clean the "family history" column
data['Do you have a family history of mental illness?'] = (
    data['Do you have a family history of mental illness?']
    .str.strip()
    .str.capitalize()
)

# Drop rows with missing family history information
data = data.dropna(subset=['Do you have a family history of mental illness?'])

# Filter for only "Yes" or "No" responses
valid_responses = ['Yes', 'No']
data = data[data['Do you have a family history of mental illness?'].isin(valid_responses)]

# Group by response and calculate counts
family_history = data['Do you have a family history of mental illness?'].value_counts()

# Calculate percentages
total_count = family_history.sum()
history_percentages = (family_history / total_count) * 100

# Print family history percentages
print("Family History with Mental Health Issues in Tech Industry:")
print(history_percentages)

# Prepare data for plotting
pivot_table = family_history.reset_index()
pivot_table.columns = ['Family History', 'Count']

# Visualize the results using a bar plot
plt.figure(figsize=(8, 6))
plt.bar(pivot_table['Family History'], pivot_table['Count'], color=['blue', 'green'])
plt.title('Mental Health Disorder Responses by Family History')
plt.ylabel('Count')
plt.xlabel('Family History')
plt.xticks(rotation=0)
plt.show()

# Visualize the results using a pie plot
plt.figure(figsize=(6, 6))
plt.pie(
    family_history,
    labels=family_history.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['blue', 'green']
)
plt.title('Family History of Mental Health Issues')
plt.show()
```
